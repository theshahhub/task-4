sobel operator

```python
img = cv2.imread("C:\\Users\\aedpu\\MANIKANDAN\\xray.jpg",0)
cv2.imshow("image" ,img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img
```

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
```

```python
sobelX = cv2.Sobel(img,cv2.CV_64F,1,0)
sobelY = cv2.Sobel(img,cv2.CV_64F,0,1)
```

```python
sobelX = np.uint8(np.absolute(sobelX))
sobelY = np.uint8(np.absolute(sobelY))
```

```python
combind = cv2.bitwise_or(sobelX,sobelY)
```

```python
plt.imshow(sobelY)
```

```python
plt.imshow(sobelX)
```

laplacian operactor

```python
lab = cv2.Laplacian(img, cv2.CV_64F,ksize=3)
lab = np.uint8(np.absolute(lab))
```

Display Results

```python
titles = ['image ','sobelX','sobelY','combind','Laplacian']
image = [img,sobelX,sobelY,combind,lab]

for i in range(5):
    plt.subplot(2,3,i+1),plt.imshow(image[i])
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```

```python
gg = cv2.addWeighted(combind,0.5,lab,0.5,0)
```

```python
plt.imshow(gg)
```

TASK VEDIO 

```python
view = cv2.VideoCapture("C:\\Users\\aedpu\\MANIKANDAN\\sample.mp4",0)
```

```python
print(view.isOpened())
```

```python
while(view.isOpened()):
    ret, frame = view.read()
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    sobelX = cv2.Sobel(gray,cv2.CV_64F,1,0)
    sobelY = cv2.Sobel(gray,cv2.CV_64F,0,1)
    sobelX = np.uint8(np.absolute(sobelX))
    sobelY = np.uint8(np.absolute(sobelY))
    combind = cv2.bitwise_or(sobelX,sobelY)
    lab = cv2.Laplacian(img, cv2.CV_64F,ksize=3)
    lab = np.uint8(np.absolute(lab))
    cv2.imshow('Grey' ,gray)
    cv2.imshow('sobelX',sobelX)
    cv2.imshow('sobelY',sobelY)
    cv2.imshow('combind',combind)
    if cv2.waitKey(30) & 0xFF == ord('q'):
        break
                
view.release() 
cv2.destroyAllWindows()
```
